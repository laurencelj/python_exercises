```python
class NoughtsAndCrosses():
    player_dict = {'player1': 'X', 'player2': 'O'}
    
    def __init__(self):
        counter = 0
        self.board = [[[' '],[' '],[' ']],[[' '],[' '],[' ']],[[' '],[' '],[' ']]]
        self.counter = counter
    
  
        
    def play_game(self):
        print(self.board[0])
        print(self.board[1])
        print(self.board[2])
        while self.winner() == True:
            if self.counter%2 == 0:
                user_input = input('Player 1, where would you like to place your counter? ( in the two character format of x-coord y-coord): ')
                if (len(user_input) != 2):
                    print('That\'s not a valid move, please try again')
                else:
                    x = int(user_input[0])
                    y = int(user_input[1])
                    if self.board[2-y][x] != [' ']:
                        print('That\'s not a valid move, please try again')
                    else:
                        self.board[2-y][x] = ' X '
                        print(self.board[0])
                        print(self.board[1])
                        print(self.board[2])
                        self.counter += 1
                        if self.winner() == False:
                            print('Player 1 wins!')
                            break
            else:
                user_input = input('Player 2, where would you like to place your counter? ( in the two character format of x-coord y-coord): ')
                if (len(user_input) != 2):
                    print('That\'s not a valid move, please try again')
                else:
                    x = int(user_input[0])
                    y = int(user_input[1])
                    if self.board[2-y][x] != [' ']:
                        print('That\'s not a valid move, please try again')
                    else:
                        self.board[2-y][x] = ' O '
                        print(self.board[0])
                        print(self.board[1])
                        print(self.board[2])
                        self.counter += 1
                        if self.winner() == False:
                            print('Player 2 wins!')
                            break
                
    def winner(self):
        s = self.board
        for i in range(2):
            if s[i][0] == s[i][1] and s[i][1] == s[i][2] and s[i][0] != [' ']:
                return False
            elif s[0][i] == s[1][i] and s[1][i] == s[2][i] and s[1][i] != [' ']:
                return False
            else:
                if s[0][0] == s[1][1] and s[1][1] == s[2][2] and s[1][1] != [' ']:
                    return False
                elif s[2][0] == s[1][1] and s[1][1] == s[0][2] and s[1][1] != [' ']:
                    return False
                else:
                    return True
```

```python
class Hangman():
    def __init__(self):
        f = open('C:\data\Dictionary.txt', 'r')
        f2 = f.read()
        dictionary = f2.split('''","''')
        self.word_to_guess = random.choice(dictionary)
        self.guess_counter = 10
        self.character_list = []
        self.guess_list = []
        self.guess_board = [' ']*len(self.word_to_guess)
        for i in range(len(str(self.word_to_guess))):
            self.character_list.append(str(self.word_to_guess)[i])
        
        
    def BeginGame(self):
        print('Welcome to hangman. Here is the board...')
        print(self.guess_board)
        while self.guess_counter > 0:
            letter = input('Please guess a letter:')
            if len(str(letter)) != 1:
                print('That\'s not a valid letter, please try again:')
            else:
                if str(letter) in self.character_list and str(letter) not in self.guess_board:
                    print(f'Well done, the letter {letter} is in the word!')
                    if ' ' not in self.guess_board:
                        print(f'YOU WON!!!! You had {self.guess_counter} guesses remaining too!')
                        break
                    else:
                        print('Here\'s how the word looks now:')
                        indices = [j for j, x in enumerate(self.character_list) if x == letter]
                        for k in indices:
                            self.guess_board[k] = str(letter)
                            print(self.guess_board)
                        
                elif str(letter) not in self.character_list:
                    if str(letter) not in self.guess_list:
                        self.guess_list.append(letter)
                        self.guess_counter -= 1                    
                        print(f'Unlucky, the letter {letter} isn\'t in the word. You have {self.guess_counter} guesses left...')
                        print('Here\'s the board again...')
                        print(self.guess_board)
                    else:
                        print(f'You\'ve already guessed the letter {letter}, have another go...')
                else:
                    print(f'You\'ve already guessed the letter {letter}, have another go...')
        else:
            print('You\'re out of guesses, bad luck. GAME OVER!')
            
           
```
